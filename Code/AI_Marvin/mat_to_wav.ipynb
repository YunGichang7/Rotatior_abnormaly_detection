### 폴더 별 파일 수 확인

```python
import os

def count_files_in_folder(folder_path):
    if not os.path.exists(folder_path):
        print("폴더가 존재하지 않습니다.")
        return

    if not os.path.isdir(folder_path):
        print("유효한 폴더 경로가 아닙니다.")
        return

    file_count = 0
    for _, _, files in os.walk(folder_path):
        file_count += len(files)
    
    return file_count

file_count = count_files_in_folder('/home/soda/Project/python/notebook/AI_Team3/Data_wav/Inner Race')
print(f"폴더 내 파일 수: {file_count}")
```

### mat to wav

```python
import os
import scipy.io as sio
import scipy.io.wavfile as wavfile
import numpy as np

def mat_to_wav(mat_file_path, output_wav_path, sample_rate=12800):
    try:
        data = sio.loadmat(mat_file_path)
        signal_data = data['signals']

        normalized_data = signal_data / np.max(np.abs(signal_data))
        wav_data = (normalized_data[0] * 32767).astype(np.int16)

        wavfile.write(output_wav_path, sample_rate, wav_data)
        print(f"{output_wav_path} 파일로 변환 완료!")
    except Exception as e:
        print(f"데이터 변환 중 오류 발생: {e}")

def convert_all_mat_to_wav(input_folder, output_folder, sample_rate=44100):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for root, _, files in os.walk(input_folder):
        for file in files:
            if file.endswith(".mat"):
                mat_file_path = os.path.join(root, file)

                relative_folder = os.path.relpath(root, input_folder)
                output_subfolder = os.path.join(output_folder, relative_folder)

                if not os.path.exists(output_subfolder):
                    os.makedirs(output_subfolder)

                output_wav_path = os.path.join(output_subfolder, file.replace(".mat", ".wav"))

                mat_to_wav(mat_file_path, output_wav_path, sample_rate)

# 사용 예시
input_folder = '/home/soda/Project/python/notebook/AI_Team3/Data'
output_folder = '/home/soda/Project/python/notebook/AI_Team3/Data_wav'
sample_rate = 12800  # 원하는 샘플 레이트로 변경할 수 있습니다.

convert_all_mat_to_wav(input_folder, output_folder, sample_rate)
```
